# Problema 3

**Construya una función “suma de gaussianas” 2-dimensional, en la forma:**

$$
f(\mathbf{x}) = -\sum_{i=1}^{k} \exp\left(-\frac{1}{2\sigma} \|\mathbf{x} - \mathbf{x}_i\|_2^2\right),
$$

donde **$x_1, x_2, ..., x_k$** son puntos en el rectángulo \[0,8] × \[0,8] elegidos de forma aleatoria (distribución uniforme). Use $k = 8$ (8 gaussianas). Aquí, $\sigma > 0$ es un parámetro de escala definido por el usuario.

**Nota:** Una vez se construyen los puntos aleatorios **$x_k$** deberá congelarlos para trabajar siempre con la misma función $f$.

Aplique varias veces alguno de los métodos implementados a la función $f$, con inicializaciones **$x_0$** distintas, de forma que se puedan obtener los diferentes mínimos locales de la función.

Elabore una tabla con todos los mínimos encontrados, y muestre visualizaciones de diferentes secuencias de aproximaciones $\{x_i\}$ convergiendo a cada uno de los mínimos locales de su $f$, al igual que en el ejercicio anterior.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
#generacion de puntos y congelacion
np.random.seed(42)

k=8
sigma = 0.5
bounds = [(0,8), (0,8)]

x_points = np.random.uniform(0, 8, (k,2))

#Funcion e implementacion
def gauss_2d_sum(x, x_points = x_points, sigma = sigma):
    #f(x) = -∑ exp(-1/(2σ) * ||x - x_i||²)
    x = np.array(x)
    if x.ndim == 1:
        x = x.reshape(1,-1)

    result = 0
    for i in range (len(x_points)):
        dist_sq = np.sum((x - x_points[i])**2, axis = -1)
        result += np.exp(-dist_sq / (2* sigma))
    return -result

#visualizacion funcion
def vis_funcion():
    x = np.linspace(0, 8, 100)
    y = np.linspace(0, 8, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i, j] = gauss_2d_sum([X[i, j], Y[i, j]])
    
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
    ax.scatter(x_points[:, 0], x_points[:, 1], 
               gauss_2d_sum(x_points), c='red', s=100, marker='o')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('f(x)')
    ax.set_title('Funcion Suma de Gaussianas 2-Dimensional')
    fig.colorbar(surf)
    plt.show()

def plot_contour():
    x = np.linspace(0, 8, 100)
    y = np.linspace(0, 8, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i, j] = gauss_2d_sum([X[i, j], Y[i, j]])
    
    plt.figure(figsize=(10, 8))
    contour = plt.contour(X, Y, Z, 20, cmap='viridis')
    plt.scatter(x_points[:, 0], x_points[:, 1], c='red', s=100, marker='o')
    plt.colorbar(contour)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Curvas de nivel de la función')
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
#optimizacion con tracking 
def optimize_tracking(x0, method='BFGS'):
    trajectory = [x0]
    
    def callback(x):
        trajectory.append(x.copy())
    
    result = minimize(gauss_2d_sum, x0, method=method, 
                     bounds=bounds, callback=callback)
    
    trajectory = np.array(trajectory)
    return result, trajectory

#MInimos locales
def find_local_minima(num_trials=50):
    minima = []
    trajectories = []
    initial_points = []
    
    for i in range(num_trials):
        #Po rendom
        x0 = np.random.uniform(0, 8, 2)
        
        # Optimizar
        result, trajectory = optimize_tracking(x0)
        
        is_new = True
        for min_point, min_value in minima:
            if np.linalg.norm(result.x - min_point) < 0.1:
                is_new = False
                break
        if is_new:
            minima.append((result.x, result.fun))
            trajectories.append(trajectory)
            initial_points.append(x0)
    
    return minima, trajectories, initial_points

#visualizacion convergencai
def convergencias_trayectorias(trajectories, minima):
    plt.figure(figsize=(12, 10))
    
    x = np.linspace(0, 8, 50)
    y = np.linspace(0, 8, 50)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i, j] = gauss_2d_sum([X[i, j], Y[i, j]])
    
    contour = plt.contour(X, Y, Z, 15, colors='gray', alpha=0.6)
    colors = plt.cm.tab10(np.linspace(0, 1, len(trajectories)))
    
    for i, (traj, color) in enumerate(zip(trajectories, colors)):
        plt.plot(traj[:, 0], traj[:, 1], 'o-', color=color, 
                linewidth=2, markersize=4, label=f'Trayectoria {i+1}')
        plt.plot(traj[0, 0], traj[0, 1], 's', color=color, 
                markersize=8, markeredgecolor='black')
        plt.plot(traj[-1, 0], traj[-1, 1], 'X', color=color, 
                markersize=10, markeredgecolor='black')
    
    #guassianas piuntos
    plt.scatter(x_points[:, 0], x_points[:, 1], c='red', 
               s=150, marker='o', edgecolors='black', label='Centros Gaussianas')
    
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Trayectorias de convergencia a minimos locales')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xlim(0, 8)
    plt.ylim(0, 8)
    plt.show()
```

```python
print("Puntos de las gaussianas (congelados)")
print(x_points)

print("\nFunción")
vis_funcion()
plot_contour()

minima, trajectories, initial_points = find_local_minima(20) #20 minimos

#Tabla
print("\nTabla de minimos locales encontrados")
results_table = []
for i, ((min_point, min_value), init_point) in enumerate(zip(minima, initial_points)):
    results_table.append({
        'Mínimo #': i+1,
        'Coordenadas': f"({min_point[0]:.4f}, {min_point[1]:.4f})",
        'Valor f(x)': f"{min_value:.6f}",
        'Punto inicial': f"({init_point[0]:.2f}, {init_point[1]:.2f})"
    })

df = pd.DataFrame(results_table)
print(df.to_string(index=False))

print("\nTrayectoroas de convergencias")
convergencias_trayectorias(trajectories, minima)

print("\nTrayectorias seleccionadas")
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()

for i, (ax, trajectory) in enumerate(zip(axes[:min(4, len(trajectories))], trajectories[:4])):
    # Crear malla para el fondo
    x = np.linspace(0, 8, 50)
    y = np.linspace(0, 8, 50)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)
    
    for j in range(X.shape[0]):
        for k in range(X.shape[1]):
            Z[j, k] = gauss_2d_sum([X[j, k], Y[j, k]])  # Corregido: gauss_2d_sum
    
    ax.contour(X, Y, Z, 10, colors='gray', alpha=0.6)
    
    ax.plot(trajectory[:, 0], trajectory[:, 1], 'o-', 
            linewidth=2, markersize=4, color='blue')
    ax.plot(trajectory[0, 0], trajectory[0, 1], 's', 
            markersize=8, color='green', label='Inicio')
    ax.plot(trajectory[-1, 0], trajectory[-1, 1], 'X', 
            markersize=10, color='red', label='Mínimo')
    
    ax.scatter(x_points[:, 0], x_points[:, 1], c='orange', 
              s=100, marker='o', alpha=0.7)
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title(f'Trayectoria {i+1}: ({trajectory[0,0]:.1f},{trajectory[0,1]:.1f}) → ({trajectory[-1,0]:.3f},{trajectory[-1,1]:.3f})')
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)

plt.tight_layout()
plt.show()
```
